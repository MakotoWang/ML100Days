```python
import numpy as np
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0, resolution="f")
#’ortho’指正射投影，後面两個参數是设置中心點

# 给整個地圖上藍色
map.drawmapboundary(fill_color = 'aqua')

# 给陸地塗上珊瑚色，湖泊塗上藍色
map.fillcontinents(color = 'coral', lake_color = 'aqua')

# 畫圖
map.drawcoastlines()

map.drawparallels(range(-90,91,30), color='#CCCCCC')  # 緯線
map.drawmeridians(range(-180,181,60), color='#CCCCCC')  # 經線

# 顯示结果
plt.show()
```
