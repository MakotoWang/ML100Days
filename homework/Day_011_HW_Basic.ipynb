### 1. 根據題目給的 DataFrame 完成下列操作：

```
- 計算每個不同種類 animal 的 age 的平均數
- 計算每個不同種類 animal 的 age 的平均數
- 將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
- 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
```


````python
# 記得先 Import 正確的套件
import numpy as np
import pandas as pd
````

````python
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
````

````python
Q1_1 = df.groupby('animal').mean()
Q1_1
````

````python
Q1_2 = df.groupby('animal').mean()
Q1_2
````

````python
Q1_3 = df.sort_values(by=['age','visits'], ascending=[True,False])
Q1_3
````

````python
df['priority'] = df['priority'].str.replace('yes','True').str.replace('no','False')
df
Q1_4 = df.astype({'priority':np.bool_})
print(Q1_4['priority'].dtypes)
Q1_4
````

### 2. 一個包含兩個欄位的 DataFrame，將每個數字減去

* 1) 該欄位的平均數
* 2) 該筆資料平均數

````python
df = pd.DataFrame(np.random.random(size=(5, 3)))
df
````

````python
df_col = df.mean(axis=0)
print(df_col)
Q2_1 = df-df_col
Q2_1
````

````python
df_row = df.mean(axis=1)
print(df_row)
Q2_2 = df-df_row
Q2_2
````

### 3. 承上題，請問：

1) 哪一比的資料總合最小
2) 哪一欄位的資料總合最小



````python
Q_sum = df.sum(axis=1)
print(Q_sum)
Q3_1 = df.sum(axis=1).argmin()
print('資料總和最小筆：', Q3_1)
````

````python
Q_sum = df.sum(axis=0)
print(Q_sum)
Q3_2 = df.sum(axis=0).argmin()
print('資料總和最小欄位：', Q3_2)
````
