```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
```

## 讀入資料

## 觀察 Age

```python
df_train = pd.read_csv("Titanic_train.csv")
df_train
```

```python
# 將性別離散轉連續
from sklearn import preprocessing
#下面兩行程式，讓 Male=1, FeMale-0.
le = preprocessing.LabelEncoder()
df_train['Sex']=le.fit_transform(df_train['Sex'])
display(df_train)
```

# Age 和 Pclass 與 Sex 的關係

```python
g = sns.catplot(x="Pclass", y="Age", hue="Sex",
               data=df_train,kind="box")
```

```python
g = sns.regplot(x="Age", y="Pclass", color="g",data=df_train)
#年齡和艙等有關連(負相關)
```

# Age 和 Sex 的關係

```python
g = sns.catplot(x="Age", y="Sex", hue="Age",
               data=df_train,kind="box")
```

```python
g = sns.regplot(x="Age", y="Sex", color="g",data=df_train)
#年齡和性別有關連(正相關)
```

```python
print('Q1：年齡和艙等有關連(負相關)，年齡和性別有關連(正相關)')
```

```python
# 計算資料點的倆倆距離
from sklearn.metrics.pairwise import nan_euclidean_distances
display(pd.DataFrame(nan_euclidean_distances(df_train[['Age','Pclass','Sex']])))
```
